```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
```

```python

# File to Load
vaccinations_data_to_load = Path("../Resources/vaccinations.csv")


# Read data set
vaccinations_data = pd.read_csv(vaccinations_data_to_load)

vaccinations_data.head()
```

```python
# Clean up data set 

cleaned_vaccinations_data = vaccinations_data[["date", "location_key", "cumulative_persons_vaccinated"]]
cleaned_vaccinations_data = cleaned_vaccinations_data.rename(columns={"location_key": "location"})


cleaned_vaccinations_data
```

```python
# Filtered location to US only
us_vaccinations_data = cleaned_vaccinations_data[cleaned_vaccinations_data["location"] == "US"]

us_vaccinations_data.head()
```
